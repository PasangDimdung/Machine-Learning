```python
import matplotlib.pyplot as plt
x1 = [1,2,3]
y1= [5,7,6]
x2 = [1,2,3]
y2= [10,14,12]
plt.plot(x1,y1, label= 'First Line')
plt.plot(x2,y2 ,label= 'Second Line')
plt.title('Example')
plt.legend()     #is used for unit or appendix or labeling box

plt.show()
```

```python
#BarChart -- Comparision
import matplotlib.pyplot as plt
x1= [2,4,6,8,10]
y1= [3,6,9,6,2]
x2= [1,3,5,7,9]
y2= [8,6,2,5,6]
plt.bar(x1,y1, label='bar1' , color='g')
plt.bar(x2,y2, label='bar2' , color='r')
plt.xlabel('plot number')
plt.ylabel('Important var')
plt.title('Barchart')
plt.legend()
plt.show()
```

```python
#Histogram --- Census, used for distribution, for large data
import numpy as np
import matplotlib.pyplot as plt
population_ages= [12,8,35,20,15,86,77,22]
bins = [0,10,20,30,40,50,60,70]
plt.hist(population_ages, bins ,rwidth= 0.8)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Population distribution')
plt.show()
```

```python
#ScatterPlot -- multiple variable correlation
import matplotlib.pyplot as plt
x= [1,2,3,4,5,6,7,8]
y= [5,2,4,2,1,4,5,2]
plt.scatter(x,y, label='scatter example',color='k', s=100, marker='.')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter')
plt.show()
```

```python
#Pie-Chart  -- slices
import matplotlib.pyplot as plt
slices= [7,2,2,13]
activities= ['sleep','eat','code','play']
cols= ['m','g','r','b']
plt.pie(slices, labels = activities, startangle=90, explode=(0,0.1,0,0), autopct='%1.1f%%')
plt.title('Pie chart')
plt.legend()
plt.show()
```
