# Matplotlib Practice

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline 
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

FUNCTIONAL METHOD

```python
plt.plot(x,y)
plt.xlabel("X-LABEL")
plt.ylabel("Y-LABEL")
plt.title('EXAMPLE-1')
plt.show()
```

OBJECT ORIENTED METHOD - The main idea in using the more formal Object Oriented method is to create figure objects and then just call methods or attributes off of that object. This approach is nicer when dealing with a canvas that has multiple plots on it. 

```python
fig = plt.figure()                        #Imagine it's like creating a blank canvas
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom, width, height (range 0 to 1)

axes.plot(x, y, 'g')
axes.set_xlabel('Set X Label')
axes.set_ylabel('Set y Label')
axes.set_title('Set Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot(x,y, 'g*--')
axes1.set_xlabel('X-LABEL')
axes1.set_ylabel('Y_LABEL')
axes1.set_title('LARGER PLOT')

axes2.plot(y,x, 'r--')
axes2.set_xlabel('x-label')
axes2.set_ylabel('y-label')
axes2.set_title('SMALLER PLOT')
```

```python
fig,axes= plt.subplots(nrows=1, ncols=2)  #.subplots() does all the things for us just specifying number of rows and colms
                                          # What it does is- creates blank canvas & add axes.
axes[0].plot(x,y, 'g*--')
axes[0].set_title('First Plot')

axes[1].plot(y,x, 'b--')
axes[1].set_title('Second plot')

plt.tight_layout()     #Incase if two plots overlap eachother
```

# Figure sizes and DPI

```python
fig,axes = plt.subplots(nrows=2, ncols=1, figsize=(8,2))

axes[0].plot(x,y)    #axes is the list of matplotlib axes so if it is list then we can index them. 

axes[1].plot(y,x)   #indexing axes 1 and using .plot() method to plot (y,x)

plt.tight_layout()
```

```python
fig.savefig('FiguresizeDPI.png', dpi=100)   #saves figure, we can set DPI also
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y**2, label='X-Squared')
ax.plot(x,y**3, label='Y-labeled')

ax.legend(loc=0)
```

# Plot Appearance

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color='purple', lw=3,ls='--',alpha=1,marker='*', markersize=20,
       markerfacecolor='Yellow', markeredgewidth=3, markeredgecolor='green')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color='purple', lw=3,ls='--')

ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
